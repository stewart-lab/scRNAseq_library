```python
import warnings

warnings.simplefilter("ignore", category=UserWarning)
```

```python
# import bone marrow data 
adata = cr.datasets.bone_marrow()
adata
```

```python
# look at the RNA velocity
## visualize the fraction of spliced/unspliced reads; these are required to estimate RNA velocity
scv.pl.proportions(adata)
```

```python
## Preprocess the data
# filter out genes which don’t have enough spliced/unspliced counts, normalize and log transform the data and restrict to the top highly variable genes. 
scv.pp.filter_and_normalize(
    adata, min_shared_counts=20, n_top_genes=2000, subset_highly_variable=False
)
# Further, compute principal components and moments for velocity estimation.
sc.tl.pca(adata)
sc.pp.neighbors(adata, n_pcs=30, n_neighbors=30, random_state=0)
scv.pp.moments(adata, n_pcs=None, n_neighbors=None)
```

```python
# Use the dynamical model from scVelo to estimate model parameters 
# this fails not sure why
scv.tl.recover_dynamics(adata, n_jobs=8)
```

```python
# run rna velocity
scv.tl.velocity(adata, mode="dynamical")
```

```python
# set up velocity kernal and compute cell-tocell transition matrix
vk = cr.kernels.VelocityKernel(adata)
vk.compute_transition_matrix()
```

```python
# visualize the velocity graph in tsne embedding
vk.plot_projection(basis="tsne")
```

```python
# explore top genes
top_genes = adata.var["fit_likelihood"].sort_values(ascending=False).index
scv.pl.scatter(adata, basis=top_genes[:10], ncols=5, frameon=False)
```

```python
## Compute pseudotime
# palantir psuedotime has previously been computed for this data set
# compute diffusion pseudotime (DPT)
# start by computing a diffusion map
sc.tl.diffmap(adata)
```

```python
# Manually supply root cell. One semi-manual way of doing this is by using extrema of diffusion components.
root_ixs = 2394  # has been found using `adata.obsm['X_diffmap'][:, 3].argmax()`
scv.pl.scatter(
    adata,
    basis="diffmap",
    c=["clusters", root_ixs],
    legend_loc="right",
    components=["2, 3"],
)

adata.uns["iroot"] = root_ixs
```

```python
adata.obsm['X_diffmap'][:, 3].argmax()
adata.uns["iroot"] = root_ixs
```

```python
# Once we found a root cell we’re happy with (a cell from the HSC cluster), we can compute DPT and compare it with the precomputed Palantir pseudotime:
sc.tl.dpt(adata)
sc.pl.embedding(
    adata,
    basis="tsne",
    color=["dpt_pseudotime", "palantir_pseudotime"],
    color_map="gnuplot2",
)
```

```python
# DPT is a bit biased towards the CLPs; it assigns very high values to that cluster which masks variation among the other states. 
# explore with violin plots to visulaize pseudotime distribution per cluster
# restrict to Monocyte or Erythroid trajectories
mono_trajectory = ["HSC_1", "HSC_2", "Precursors", "Mono_1", "Mono_2"]
ery_trajectory = ["HSC_1", "Ery_1", "Ery_2"]

# plot the Monocyte trajectory
mask = np.in1d(adata.obs["clusters"], mono_trajectory)
sc.pl.violin(
    adata[mask],
    keys=["dpt_pseudotime", "palantir_pseudotime"],
    groupby="clusters",
    rotation=-90,
    order=mono_trajectory,
)

# plot the Erythroid trajectory
mask = np.in1d(adata.obs["clusters"], ery_trajectory)
sc.pl.violin(
    adata[mask],
    keys=["dpt_pseudotime", "palantir_pseudotime"],
    groupby="clusters",
    rotation=-90,
    order=ery_trajectory,
)
```

```python
#the Palantir pseudotime appears to be less biased towards CLPs, let’s use that for the PseudotimeKernel below.

# Compute a transition matrix based on Palantir pseudotime
pk = cr.kernels.PseudotimeKernel(adata, time_key="palantir_pseudotime")
pk.compute_transition_matrix()

print(pk)
```

```python
# Visualize
# We do not make use of RNA velocity here, CellRank implements a general way of visualizing k-NN graph-based transition matrices via streamlines in any embedding. 
# Thus, the dynamics in the following plot are purely informed by the pseudotime and the k-NN graph, and not by RNA velocity.
pk.plot_projection(basis="tsne", recompute=True)
```
